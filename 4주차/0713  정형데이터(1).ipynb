## 1. 빅데이터의 이해

### 1.1 빅데이터의 개념

- 일반적인 데이터베이스 소프트웨어가 데이터 수집 , 저장 , 관리 , 분석할 수 있는 역 량을 넘어서는 규모의 데이터
- **3V**
    - Volumne : 데이터의 규모 / 양
    - Variety : 데이터의 종류 / 유형
    - Velocity : 데이터의 생성 / 처리 속도

### 1.2 빅데이터의 중요성

- 스마트폰 , 스마트패드 등 모바일 장치의 확산
- 트위터 , 페이스북 등 소셜미디어의 성장
- RFID, IoT 의 이용 확대
- 저장장치 비용 하락
- 클라우드 컴퓨팅 확산

### 1.3 빅데이터의 처리방식

> #### 데이터 종류
- 정형 : 고정된 필드에 저장된 데이터
    - ex) 관계형 데이터베이스, 스프레드시트
- 반정형 : 고정된 필드에 저장되어 있지는 않지만, 메타데이터나 스키마 등을 포함하는 데이터
    - ex) XML, HTML, 텍스트
- 비정형 : 고정된 필드에 저장되어 있지 않은 데이터
    - ex) 텍스트 분석이 가능한 텍스트 문서, 이미지, 동영상, 음성 데이터
    
- 데이터 수집
- 데이터 저장(Hadoop / RDBMS)
- 데이터 처리(Spark)
- 데이터 분석

#### 분산 / 병렬 처리
- 하둡
    - 분산 데이터 인프라
    - 맵 리듀스
- 스파크
    - 분산된 데이터 처리 도구
    - 속도 빠름 / 실시간 처리 적합
    
## 2. DBMS

### 2.1 데이터베이스 개요

- 80% 비정형 / 20% 정형

> #### 데이터베이스 조건
- 통합(Integrated)
- 저장(Stored)
- 공유(Shared)
- 운영(Operational)

> #### 데이터베이스 특징
- 실시간 접근성
- 계속적인 변화
- 동시 공유 가능
- 내용에 의한 참조 

> #### DBMS의 역할
- create databases
- insert, update and delete data
- sort and query data
- create form and report

> #### DBMS 역할과 장점
- Improved data sharing
- Improved data security
- Minimized data inconsistency
- Improved data access
- Improved decision making
- Increased end user productivity
- Reduce application development time

> #### DBMS 종류
- 계층형 모델
- 네트워크형 모델
- 관계 데이터 모델
- 객체 관계 모델

## GIt

- History 관리하는 도구

#### Git 필수 명령
- add : 커밋할 목록에 추가
- commit : 커밋(히스토리의 한단위) 만들기
- push : 현재까지 역사(commits) Github에 밀어넣기

#### 실습
```shell
git init # ls-A 명령어로 생성된 .git 폴더 확인
git add report_card.pdf # 커밋할 목록에 추가 add
git commit -m "report card: Add question PDF" # 한 단위 생성
git commit -sm "report card: Show the sum of each grade" # -s 옵션으로 서명을 포함
git shortlog # 커밋할 목록 확인

git status # 상태 확인
git diff # 변경 내용 확인
git log # commit한 내용 확인

# Github 원격저장소 등록
git remote add origin https://github.com/sunnight9507/Git-test.git 

git push origin master # 원격저장소(origin)에 push

git commit --amend # 가장 위에 있는 commit 수정
# vi 에디터는 i, a를 눌러 수정가능
# :wq 명령어를 입력해 나갈 수 있다.

git push origin master # 바로 push 할 경우 오류 발생

git push origin master --force # 강제로 push / --force 또는 -f 옵션 사용 가능

touch test; git add test # 빈 파일 생성 후 add
git reset # reset으로 add한 것 취소
git status # 상태 확인

git add test; git commit -sm “test”# commit
git push origin master # push
git reset HEAD~1 # 가장 최근 commit 삭제
git push origin master --force # 강제로 Github 에 있는 tree 도 밀어넣어서 수정

git clone https://github.com/Taeung/git-training.git # clone으로 fork한 repo 받아오기
cd git-training # clone한 프로젝트 폴더로 이동
git checkout -b develop # pull-request 작업할 브렌치 생성 / # 같은 폴더에 또 다른 세상을 만든다는 느낌
cd pull_reqeust_test # pull_request_test로 이동
mkdir sun; cd sun # 폴더 생성 후 이동
# 해당 폴더에 아무 파일 복사

git add test # 추가한 폴도 통째로 add
git commit -sm “test pull request” # 준비된 파일들 commit
git push origin develop # fork한 repo의 develop 브렌치로 push

# pull-request 진행

# rebase
git remote add upstream https://github.com/Taeung/git-training.git
git fetch upstream dev # upstream의 dev 브렌치 가져오기
git status # 현재 내 브렌치가 develop인지 확인
git rebase upstream/dev # rebase
```
