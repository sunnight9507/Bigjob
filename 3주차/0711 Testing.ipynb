> ### assert
- 자동화 test의 기본
- function마다 assert를 넣어 체크 가능
- testing하기 위해 직접 코드를 작성
- ```Python
    def test_sum():
        assert sum([1,2,2]) == 6, "error"
    test_sum() # error```

> ### Unit Test
- unittest.TestCase이 가장 범용적임 => 다양한 Testing 가능
- ```Python
    %%writefile moon.py # file test
    import unittest
    class TestSum(unittest.TestCase):
        def test_sum(self):
            self.assertEqual(sum([1,2,3]), 6) # assert sum([1,2,3]) == 6
    if __name__ == '__main__':
        unittest.main()```
- ```Python
    import sys # jupyter notebook version
    class TestSum(unittest.TestCase):
        def test_sum(self):
            self.assertEqual(sum([1,2,3]), 6,'error') # assert sum([1,2,3]) == 6
    s = unittest.TestLoader().loadTestsFromTestCase(TestSum)
    unittest.TextTestRunner(verbosity=2, # 설명하는 정도
                           ).run(s)```
- Example
- ```Python
    class Book:
        def __init__(self, name):
            self.name = name
            self._genre = 'romantic'
            self._world = 'dark'
        def can_read(self, person):
            if person.role == 'can':
                return True
            else:
                return False
    class Person:
        def __init__(self, role):
            self.role = role
    class TestBook(unittest.TestCase):
        def setUp(self): # overriding
            print('Start')
            self.book = Book('SSS')
        def test_can_read(self):
            print('test1')
            person = Person('can1')
            self.assertFalse(self.book.can_read(person))
        def tearDown(self):
            print('End')
    s = unittest.TestLoader().loadTestsFromTestCase(TestBook)
    unittest.TextTestRunner(verbosity=2).run(s)```
