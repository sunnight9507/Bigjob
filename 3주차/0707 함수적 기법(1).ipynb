    - 함수 정의 확인할 때 **`*`** 기준으로 **`오른쪽`**에 있는 값은 **keword만 지원**

#### parameter vs argument
- argument(전달인자)
    - 1) keword
        - 값을 지정해주는 방식 ex) func(a=1, b=2, c=3)
        - 위치 변경 가능
        - default 값 정할 수 있다.(주의 : 마지막 변수부터 값을 지정해주어야 한다.)
            - ```Python 
                def func(a, b, c=3):
                    return```
    - 2) positioning
        - 위치에 따라 값을 넣어줌
    - 식 입력 가능
        - ```Python
            1 if True else 2```
- parameter(매개변수)
    - 1) 함수 정의에 **`/`**, **`*`**가 없는 경우 위치 또는 키워드로 함수에 전달 가능
    - 2) **`/`** 기준으로 **`왼쪽`**에 있는 값은 **positioning**
    - 3) **`*`** 기준으로 **`오른쪽`**에 있는 값은 **keword**
    - 4) `def func(*a): return a` : 가변 포지션 / a는 tuple / func(1,2,3)
    - 5) `def func(**a): return a` : 가변 키워드 / a는 dictionary / func(a=1, b=2) / 관례상 \*\*kw라고 함

#### iterable
- **'\_\_iter\_\_'**가 있을 경우 for문 젹용 가능
- **next 호출 가능** : next를 호출할 때 메모리가 올라가므로 메모리 절약 가능
    - ```Python
        a = [1,2,3]; b = iter(a); next(b)```
- sequence 개념을 잃어버림
    - error
    - ```Python
        a = [1,2,3]; b = iter(a); b[0]```
    - ```Python
        a = [1,2,3]; b = iter(a); 
        for i in b: 
            print(i); 
            break;```
- iterator vs iterable
    - '\_\_next\_\_'
    - ```Python
        from collections.abc import Iterable, Iterator 
        set(dir(Iterator)) - set(dir(Iterable))```
- **enumerate**
    - ```Python 
        for i in enumerate([1,2,3]): 
            print(i)```
- **all** : 모든 value가 True인지
- **any** : 모든 value 중 하나 이상의 value가 True인지
    
#### generater
- **next 호출 가능**
- **tuple로 이루어진 for문**
    - ex) `(i for i in range(10))`
- **yield**
    - ```Python 
        def func(): 
            yield from range(10)
        data = func()
        next(data)```

#### else
- **1) if - else**
- **2) 반복문 끝난 것 확인**
    - ```Python 
        for i in range(3): 
            print(i); 
        else: 
            print('end')```
    - break 구문은 else 구문 실행되지 않음
- **3) try문에서 error가 발생하지 않을 때**
    - ```Python
        try: a = 1/0; 
        except: print('error'); 
        else: print('no error')```

#### higher order function
- 함수를 인자로 받고, 함수를 리턴할 수 있는 함수
- **map**
    - ```Python 
        list(map(x, [1,2,3]))
        list(map(lambda x,y:x+y, range(10), range(10)))```
    - x에 [1,2,3]을 mapping
    - iter 방식으로 인자 하나씩 next로 호출
    - 속도가 빠르다.
- **filter**
    - ```Python
        filter(lambda x:x>3, [1,2,3,4,5])```
    - **predicate** : 상태 반환하는 함수
        - True, False 반환
    - True 값만 반환
- **reduce**
    - ```Python
        from functools import reduce 
        reduce(lambda x,y : x+y, [1,2,3,4,5])```

#### partial
- 원하는 함수를 변형 가능
    - ```Python
        def add(x,y): 
            return x + y; 
        partial(add, y = 1)```
