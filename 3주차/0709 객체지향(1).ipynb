> ### Class
- **class attribute** : 클래스가 접근할 수 있는 것
    - **`dir(Class)`**를 통해 확인 가능
    - class내에 없는 값 입력할 경우 AttributeError 발생
    - 값 변경 가능
        - ```Python
        class A:
            a = 1
        A.a = 3
        A.a # 3```
- **instance attribute** : **각 객체별 고유값**
    - 값 유무 확인 : **`vars(object)`** or **`object.__dict__`**
    - instance attribute 값이 존재하지 않을 경우 class attribute 값 출력
- 사용하기 위해 **instance화** 해야 한다.
    - **`__init__`**이 실행이 되고 **instance attribute가 생성 됨**
    - ```Python
        class A:
            a = 1
            b = 2
            def __init__(self):
                self.a = 3
        a = A()
        a.a # 3```
- **값의 추가, 삭제 가능**
    - class 변수 추가, 삭제 가능
- type casting을 instance화로 사용
    - ```Python
        a = int('a')
        list(map(lambda x:x+1, [1,2,3,]))```
- class로 접근할 경우 function / instance로 접근할 경우 method
- **classmethod**
    - class가 method 접근할 때 사용
    - instance가 접근할 때 class로 인식
    - ```Python
        class B:
            x= 1
            def __init__(self):
                print('init')
            @classmethod # 관례상 cls
            def t(cls):
                print('t')
        B.t() # t```
- **staticmethod**
    - Class, instance 변수로 접근 제한
    - 거의 사용하지 않음
- `type`개념과 동일
    - ```Python
        type(class)
        type(type)
        class.__class__
        type.__class__```
- **Composition**
    - 다른 Class를 인자로 받아 새로 초기화   
- **상속**
    - ex) `class T(상속받을 class):`
    - 상속 순서 확인 방법
        - ```Python
            class.__MRO__ # tuple
            class.mro() # list```
        - ```Python
            class A:
                def __init__(self):
                    self.x = 'A'
                    print('A')
            class B(A):
                def __init__(self):
                    self.x = 'B'
                    A.__init__(self)
                    print('B')
            class C(A):
                def __init__(self):
                    self.x = 'C'
                    A.__init__(self)
                    print('C')
            class D(B,C):
                def __init__(self):
                    B.__init__(self)
                    C.__init__(self)
                    print('D')
            d = D() # ACABD / A가 중복으로 나옴```
        - **`super()`**
            - 부모에 있는 것을 사용하기 위해 쓰임
            - 중복 문제 해결
        - ```Python
        class A:
            def __init__(self):
                self.x = 'A'
                print('A')
        class B(A):
            def __init__(self):
                self.x = 'B'
                super().__init__()
                print('B')
        class C(A):
            def __init__(self):
                self.x = 'C'
                super().__init__() # python 3.x version
                print('C')
        class D(B,C):
            def __init__(self):
                super(D, self).__init__() # python 2.x version
                print('D')
        d = D() # ACBD ```
        - 다중 상속 받을 때 **부모를 독립적으로 만들 때 `Mixin` 사용**
