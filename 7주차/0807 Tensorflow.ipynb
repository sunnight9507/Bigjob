```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(iris.iloc[:,:-1], iris.species)
```

```python
knn.predict([[3,3,3,3]])
```

```python
ohe = pd.get_dummies(iris.species)
iris = pd.concat([iris, ohe], axis = 1)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, input_shape=(4,), activation='relu'), # tf.keras.activations.relu / tf.nn.relu
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_fn():
    model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, input_shape=(4,), activation='relu'), # tf.keras.activations.relu / tf.nn.relu
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
    ])
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    return model
```

```python
kc = KerasClassifier(build_fn, epochs = 10)
```

```python
kc.fit(iris.iloc[:,:-3], iris.iloc[:,-3:])
```

```python
from tensorflow.keras.callbacks import Callback, TensorBoard
```

```python
class MyCallBack(Callback):
    def on_epoch_begin(self, epoch, logs=None):
        print(epoch, 'start!!!!!!!!!')
```

```python
def build_fn():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(16, input_shape=(28*28,), activation='relu'), # tf.keras.activations.relu / tf.nn.relu
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    return model
```

```python
model = build_fn()
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(-1,28*28)

model.fit(X_train, y_train, epochs=5, callbacks=[MyCallBack()])
```

```python
model.fit(X_train, y_train, epochs=5, callbacks=[TensorBoard()])
```
