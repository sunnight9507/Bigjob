#### Copy

```python
a = b = [1,2,3]
a[0] = 100
b
```

```python
a = [1,2,3]
b = a.copy()
a[0] = 100
b
```

```python
a = [1,2,3]
b = a[:] # copy와 동일한 개념
a[0] = 100
b
```

```python
a = [[1,2],[3,4]]
b = a.copy()
a[0][0] = 100 # 겹으로 쌓여져 있을 경우 copy 안됨
b
```

```python
import copy

a = [[1,2],[3,4]]
b = copy.deepcopy(a)
a[0][0] = 100
b
```

```python
y = x = np.array([1,2,3])
y[0] = 100
x 
```

```python
x = np.array([[1,2],[3,4]])
y = x.view() # Numpy view : Python의 copy와 동일
x[0,0] = 100
y
```

```python
x = np.array([[1,2],[3,4]])
y = x.copy() # Numpy copy : Python의 deepcopy와 동일
x[0,0] = 100
y
```
